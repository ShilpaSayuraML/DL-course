```python
import numpy as np
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow

# Create a blank image
blank = np.zeros((500,500), dtype='uint8')

# Draw a rectangle and a circle on the blank image
rectangle = cv2.rectangle(blank.copy(), (30,30), (400,400), 255, -1)
circle = cv2.circle(blank.copy(), (215,215), 150, 255, -1)

# Display the shapes
cv2_imshow(rectangle)
cv2_imshow(circle)
print()
# Bitwise AND --> intersecting regions
bitwise_and = cv2.bitwise_and(rectangle, circle)
cv2_imshow(bitwise_and)
print()
# Bitwise OR --> non-intersecting and intersecting regions
bitwise_or = cv2.bitwise_or(rectangle, circle)
cv2_imshow(bitwise_or)
print()
# Bitwise XOR --> non-intersecting regions
bitwise_xor = cv2.bitwise_xor(rectangle, circle)
cv2_imshow(bitwise_xor)
print()
# Bitwise NOT
bitwise_not = cv2.bitwise_not(circle)
cv2_imshow(bitwise_not)
```

```python
import cv2 as cv
from google.colab import files
from google.colab.patches import cv2_imshow

# Upload the image
uploaded = files.upload()

# Load the uploaded image
image_path = next(iter(uploaded))
print(image_path)
img = cv2.imread(image_path)

cv2_imshow(img)
print()

# BGR to Grayscale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
print()
# BGR to HSV
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
cv2_imshow(hsv)

# BGR to L*a*b
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)
cv2_imshow(lab)
print()

# BGR to RGB
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv2_imshow(rgb)
print()
# HSV to BGR
lab_bgr = cv.cvtColor(lab, cv.COLOR_LAB2BGR)
cv2_imshow(lab_bgr)

```

```python
import matplotlib.pyplot as plt

blank = np.zeros(img.shape[:2], dtype='uint8')

mask = cv.circle(blank, (img.shape[1]//2,img.shape[0]//2), 100, 255, -1)

masked = cv.bitwise_and(img,img,mask=mask)
cv2_imshow(masked)

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(gray)

# GRayscale histogram
gray_hist = cv.calcHist([gray], [0], mask, [256], [0,256] )

# plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
plt.plot(gray_hist)
plt.xlim([0,256])
plt.show()

# Colour Histogram

plt.figure()
plt.title('Colour Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
colors = ('b', 'g', 'r')
for i,col in enumerate(colors):
    hist = cv.calcHist([img], [i], mask, [256], [0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])

plt.show()
```
