# People or Pedistrian Detection

```python
#download files

import numpy as np
import cv2
import urllib.request

```

```python
url1='https://raw.githubusercontent.com/PacktPublishing/Hands-On-Computer-Vision-with-OpenCV-4-Keras-and-TensorFlow-2/master/Section%203/Haarcascades/haarcascade_fullbody.xml'

urllib.request.urlretrieve(url1, 'haarcascade_fullbody.xml')

url2='https://raw.githubusercontent.com/PacktPublishing/Hands-On-Computer-Vision-with-OpenCV-4-Keras-and-TensorFlow-2/master/Section%203/images/walking.avi'

urllib.request.urlretrieve(url2, 'walking.avi')
```

```python
import cv2
import numpy as np

from google.colab.patches import cv2_imshow


# Create our body classifier
body_classifier = cv2.CascadeClassifier('/content/haarcascade_fullbody.xml')

# Initiate video capture for video file
cap = cv2.VideoCapture('/content/walking.avi')

# Loop once video is successfully loaded
while cap.isOpened():

    # Read first frame
    ret, frame = cap.read()
    frame = cv2.resize(frame, None,fx=0.5, fy=0.5, interpolation = cv2.INTER_LINEAR)

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Pass frame to our body classifier
    bodies = body_classifier.detectMultiScale(gray, 1.2, 3)

    # Extract bounding boxes for any bodies identified
    for (x,y,w,h) in bodies:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
        cv2_imshow(frame)

    if cv2.waitKey(1) == 13: #13 is the Enter Key
        break

cap.release()
cv2.destroyAllWindows()
```

## Car or Vehicle Detection


```python
url1='https://raw.githubusercontent.com/PacktPublishing/Hands-On-Computer-Vision-with-OpenCV-4-Keras-and-TensorFlow-2/master/Section%203/Haarcascades/haarcascade_fullbody.xml'

urllib.request.urlretrieve(url1, 'haarcascade_car.xml')

url2='https://raw.githubusercontent.com/PacktPublishing/Hands-On-Computer-Vision-with-OpenCV-4-Keras-and-TensorFlow-2/master/Section%203/images/cars.avi'

urllib.request.urlretrieve(url2, 'cars.avi')
```

```python
import cv2
import time
import numpy as np

from google.colab.patches import cv2_imshow

# Create our body classifier
car_classifier = cv2.CascadeClassifier('/content/haarcascade_car.xml')

# Initiate video capture for video file
cap = cv2.VideoCapture('/content/cars.avi')


# Loop once video is successfully loaded
while cap.isOpened():

    time.sleep(.05)
    # Read first frame
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Pass frame to our car classifier
    cars = car_classifier.detectMultiScale(gray, 1.4, 2)

    # Extract bounding boxes for any bodies identified
    for (x,y,w,h) in cars:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
        cv2_imshow(frame)

    if cv2.waitKey(1) == 13: #13 is the Enter Key
        break

cap.release()
cv2.destroyAllWindows()
```
