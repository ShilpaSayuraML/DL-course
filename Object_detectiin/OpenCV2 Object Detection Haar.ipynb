###Recognition

```python
import cv2
from matplotlib import pyplot as plt
from PIL import Image
import urllib.request
import numpy as np

# Opening image
img = cv2.imread("image.jpg")

img_rgb = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)

# Convert the image to grayscale
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)

# Load the cascade
stop_data = cv2.CascadeClassifier('/content/stop_data.xml')

# Detect stop signs
found = stop_data.detectMultiScale(img_gray, minSize=(20, 20))

# Draw rectangles around the detected stop signs
for (x, y, width, height) in found:
    cv2.rectangle(img_rgb, (x, y), (x + height, y + width), (0, 255, 0), 5)

# Display the result
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

###Haar Cascade classifier for face detection

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

# Load the image
img = cv2.imread("/content/person2.jpg")

# Convert the image to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Load the Haar Cascade classifier for face detection
face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
# Detect faces in the image
faces = face_cascade.detectMultiScale(img_gray, scaleFactor=1.3, minNeighbors=5)

# Draw rectangles around the detected faces
for (x, y, width, height) in faces:
    cv2.rectangle(img, (x, y), (x + width, y + height), (0, 255, 0), 5)

# Display the result
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import urllib.request
import cv2
import numpy as np
from PIL import Image

# URL of the Haar Cascade file
cascade_url = "https://raw.githubusercontent.com/informramiz/Face-Detection-OpenCV/master/data/haarcascade_frontalface_alt.xml"

# Download the Haar Cascade file
urllib.request.urlretrieve(cascade_url, "haarcascade.xml")

# Load the Haar Cascade classifier
face_cascade = cv2.CascadeClassifier("haarcascade.xml")

# Example usage: Load an image
image_url = "https://raw.githubusercontent.com/informramiz/Face-Detection-OpenCV/master/data/test1.jpg"
image_path, _ = urllib.request.urlretrieve(image_url, "imagex.jpg")
img = cv2.imread("imagex.jpg")

# Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

# Draw rectangles around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Display the image with rectangles around faces
Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

```
