###Load video from url, save locally and display frame by frame

```python

import cv2
import urllib.request
from google.colab.patches import cv2_imshow

# Function to load video from URL and save locally
def download_and_save_video(url, filename='video.mp4'):
    urllib.request.urlretrieve(url, filename)

# Function to load video from local file and display it
def load_and_display_video(filename='video.mp4'):
    cap = cv2.VideoCapture(filename)

    i=0
    while(cap.isOpened()):
        ret, frame = cap.read()
        if not ret:
            break
        i=i+1
        print('frame ', i)
        # Display the frame using cv2_imshow
        cv2_imshow(frame)

        # Break the loop when 'q' key is pressed
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break

    cap.release()

# Example usage:
video_url = 'https://github.com/intel-iot-devkit/sample-videos/raw/master/car-detection.mp4'

download_and_save_video(video_url,'video.mp4')
print('video.mp4', 'saved')
# Load and display the locally saved video
load_and_display_video('video.mp4')
```

### Display saved video

```python
from IPython.display import HTML
from base64 import b64encode

# Function to display video in Colab
def display_video(file_path='video.mp4'):
    video_file = open(file_path, "rb").read()
    video_encoded = b64encode(video_file).decode('ascii')
    video_tag = f'<video controls src="data:video/mp4;base64,{video_encoded}">'
    display(HTML(video_tag))

# Example usage:
video_path = 'video.mp4'
display_video(video_path)
```
