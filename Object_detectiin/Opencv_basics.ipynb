```python
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow  # Import cv2_imshow for displaying images

# Upload the image
uploaded = files.upload()

# Load the uploaded image
image_path = next(iter(uploaded))
print(image_path)
img_uploaded = cv2.imread(image_path)

cv2_imshow(img_uploaded)

# Get the original width and height of the image
height, width = img_uploaded.shape[:2]

# Calculate the scale factor for resizing to width 200 while preserving aspect ratio
scale_factor = 300 / width

# Resize the image
img = cv2.resize(img_uploaded, (int(width * scale_factor), int(height * scale_factor)))

# Display the resized image

# Display the original image
cv2_imshow(img)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

# Blur
blur = cv2.GaussianBlur(img, (7, 7), cv2.BORDER_DEFAULT)
cv2_imshow(blur)

# Edge Cascade
canny = cv2.Canny(blur, 125, 175)
cv2_imshow(canny)

# Dilating the image
dilated = cv2.dilate(canny, (7, 7), iterations=3)
cv2_imshow(dilated)

# Eroding
eroded = cv2.erode(dilated, (7, 7), iterations=3)
cv2_imshow(eroded)

# Resize
resized = cv2.resize(img, (100, 100), interpolation=cv2.INTER_CUBIC)
cv2_imshow(resized)

# Cropping
# Crop the region with coordinates (59, 50) and width 60, height 60
x, y, w, h = 59, 50, 60, 60
cropped_img = img[y:y+h, x:x+w]

# Display the cropped image
cv2_imshow(cropped_img)
```

```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow  # Import cv2_imshow for displaying images

# Upload the image
uploaded = files.upload()

# Load the uploaded image
image_path = next(iter(uploaded))
print(image_path)

# Read the image
img = cv2.imread(image_path)
cv2_imshow(img)

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Perform edge detection using Canny
canny = cv2.Canny(gray, 125, 175)

# Find contours
contours, hierarchies = cv2.findContours(canny, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
print(f'{len(contours)} contour(s) found!')

# Create a blank image to draw contours on
blank = np.zeros_like(img)

# Draw contours on the blank image
cv2.drawContours(blank, contours, -1, (0, 0, 255), 1)
cv2_imshow( blank)
```

```python

import cv2 as cv
import numpy as np

blank = np.zeros((500,400,3), dtype='uint8')
cv2_imshow(blank)
print()
# 1. Paint the image a certain colour
blank[100:100, 200:200] = 0,0,255
cv2_imshow(blank)
print()
# 2. Draw a Rectangle
cv.rectangle(blank, (0,0), (blank.shape[1]//2, blank.shape[0]//2), (0,255,0), thickness=-1)
cv2_imshow(blank)
print()
# 3. Draw A circle
cv.circle(blank, (blank.shape[1]//2, blank.shape[0]//2), 40, (0,0,255), thickness=-1)
cv2_imshow(blank)
print()
# 4. Draw a line
cv.line(blank, (100,150), (200,300), (255,255,255), thickness=3)
cv2_imshow(blank)
print()
# 5. Write text
position = (50, 350)  # Position where the text will be drawn (x, y)
cv.putText(blank, 'My name is Anura!', position , cv.FONT_HERSHEY_TRIPLEX, 1.0, (255,255,255), 2)
cv2_imshow(blank)
```

```python
def rotate_image(image, angle):
    # Get image dimensions
    (height, width) = image.shape[:2]
    # Calculate the rotation matrix
    rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1.0)
    # Apply the rotation to the image
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
    return rotated_image

# Rotate the image by 45 degrees clockwise
rotated_img = rotate_image(img, -45)
cv2_imshow(rotated_img)
```

```python
# Flipping
flip = cv2.flip(rotated_img, -1)
cv2_imshow(flip)
```
