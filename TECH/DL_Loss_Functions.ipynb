# Loss functions
They tell how different  the predicted output from  the actual output.

For regression tasks we use :
* Mean Square Error
* Root Mean Square Error
* Mean Absolute Error

For classification tasks we use:
* Cross-Entropy Loss


MSE - Mean Squred Error

Mean-square-error.png

```python
import numpy as np

act = np.array([1, 2, 1.9])
pred = np.array([1,2.2,1.7])

def mean_squared_error(act, pred):

   diff = pred - act
   print(pred, act, diff)
   differences_squared = diff ** 2
   print(differences_squared)
   mean_diff = differences_squared.mean()

   return mean_diff

print(mean_squared_error(act,pred))
```

## Using Sci-kit learn

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(act, pred)
print(mse)
```

Root Mean Square Error (RMSE)

Mean-square-error.pngRMSE.png


```python
import numpy as np
def root_mean_squared_error(act, pred):

   diff = pred - act
   differences_squared = diff ** 2
   mean_diff = differences_squared.mean()
   rmse_val = np.sqrt(mean_diff)
   return rmse_val



print(root_mean_squared_error(act,pred))
```

```python

from sklearn.metrics import mean_squared_error
mean_squared_error(act, pred, squared = False)
```

## Mean Absolute Error (MAE)

MAE.gif

```python
import numpy as np
def mean_absolute_error(act, pred):
    diff = pred - act
    print(pred, act, diff)
    abs_diff = np.absolute(diff)
    print(abs_diff)
    mean_diff = abs_diff.mean()
    return mean_diff

mae=mean_absolute_error(act,pred)
print(mae)
```

```python
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(act,pred)
print(mae)

```

```python
import numpy as np

#Sigmoid function

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

# yHat is predicted value / probability value
# output of  hypothesis / sigmoid function
 # y represents the actual label

# Value of cost function when y = 1 is -log(h(x))
# Value of cost function when y = 0 is -log(1 - h(x))

def cross_entropy_loss(yHat, y):
    if y == 1:
      return -np.log(yHat)
    else:
      return -np.log(1 - yHat)

#generates values from -3 to 3 with a step size of 0.1.
z= np.arange(-2, 2, 0.1)
print(z)
print()
h_z = sigmoid(z)
print(h_z)
print()

#cost when 'y' is 1 (positive class).
cost_1 = cross_entropy_loss(h_z, 1)

#cost when 'y' is 0 (positive class).
cost_0 = cross_entropy_loss(h_z, 0)
print(cost_1)
print()
print(cost_0)

```

plot cross entropy

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(8,6))
plt.plot(h_z, cost_1, label='J(w) if y=1')
plt.plot(h_z, cost_0, label='J(w) if y=0')
plt.xlabel('$\phi$(z)')
plt.ylabel('J(w)')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
